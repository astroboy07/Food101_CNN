# Preprocessing the data

Since the pixel values are not scaled and in `uint` data type, we create a preprocessing function which scales and change the datatype. 

```python
def preprocessing_image(image, label, image_shape=224):
  image = tf.image.resize(images=image,
                          size=[image_shape, image_shape])
  image = tf.cast(image, dtype=tf.float32)
  return image, label
```

```python
preprocessed_image = preprocessing_image(image, label)
# check the shape of the image
preprocessed_image[0].shape, image.shape
```

```python
preprocessed_image[0].numpy(), image.numpy()
```

Before we use the `train_data` inside our model we need to batch and preprocess it. By preprocessing I mean a couple of specific things which are as follows:

* `map()`: maps the function to a target dataset (`preprocessing_image()` --> `train_data` and `test_data`)

* `shuffle()`: shuffle the elements of a target dataset using the `buffer_size`; which is the size of the data (ideally). However, this will have implications on memory so we choose 1000 as an abstract value.

* `batch()`: turns target dataset in to batches.

* `prefetch()`: prepares the subsequent batches of data on CPU as the model trains on previous batch of data on GPU. We can further speed up by using multithreading loading.

**Workflow**:

Target dataset --> map() --> shuffle() --> batch() --> prefetch() --> Prefetched dataset



```python
# Preprocess the target datasets using the above prescription
train_data = train_data.map(map_func=preprocessing_image, num_parallel_calls=tf.data.AUTOTUNE)
train_data = train_data.shuffle(buffer_size=1000).batch(batch_size=32).prefetch(buffer_size=tf.data.AUTOTUNE)

test_data = test_data.map(map_func=preprocessing_image, num_parallel_calls=tf.data.AUTOTUNE)
test_data = test_data.batch(batch_size=32).prefetch(buffer_size=tf.data.AUTOTUNE) # don't need to shuffle as non-shuffled test data is requied for evaluation 
```

# Creating modelling callbacks 

```python
# Creating TensoBoard callback
import datetime

def create_tensorboard_callback(dir_name, experiment_name):
  log_dir = dir_name + "/" + experiment_name + "/" + datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
  tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
  print(f"Saving TensorBoard logfile to: {log_dir}")
  return tensorboard_callback
```

```python
checkpoint_path = "model_checkpoints/cp.ckpt" 
model_checkpoint = tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                      monitor="val_accuracy", 
                                                      save_best_only=True, 
                                                      save_weights_only=True, 
                                                      verbose=0) 
```

# Modeling

```python
# Set up mixed precision training
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy(policy='mixed_float16')
mixed_precision.global_policy()
```

### Build a feature extraction model using mixed precision training

```python
from tensorflow.keras import layers, Model
from tensorflow.keras.applications import EfficientNetB1

# Create base model
input_shape = (224, 224, 3)
base_model = EfficientNetB1(include_top=False)
base_model.trainable = False

# Create DNN with base model
inputs = layers.Input(shape=input_shape, name='input_layer')
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D(name='GlobalAveragePooling2D')(x)
x = layers.Dense(len(class_names))(x)
# Separate activation of output layer so we can output float32 activations
outputs = layers.Activation('softmax', dtype=tf.float32, name='output_layer')(x)

model_feature_extract = Model(inputs, outputs)

# Compile the model
model_feature_extract.compile(loss=tf.keras.losses.sparse_categorical_crossentropy,
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])
```

```python
model_feature_extract.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model_feature_extract,
           show_shapes=True,
           show_dtype=True)
```

```python
# Diving deep into the layers to check the dtypes
for layer in model_feature_extract.layers[1].layers[:10]:
  print(layer.name, layer.trainable, layer.dtype, layer.dtype_policy)

# Just a note: layer.dtype: data type the layer stores its variable in, layer.dtype_policy: data type the layer computes its variable in 
```

Fit the feature extraction model

```python
history_feature_extract = model_feature_extract.fit(train_data,
                                                    epochs=5,
                                                    steps_per_epoch=len(train_data),
                                                    validation_data=test_data,
                                                    validation_steps=int(0.15 * len(test_data)),
                                                    callbacks=[create_tensorboard_callback("training_logs",
                                                                                           "efficientnetb1_feature_extract"),
                                                               model_checkpoint])
```

```python
# Evaluate model on whole test dataset
evaluation_feature_extract = model_feature_extract.evaluate(test_data)
evaluation_feature_extract
```

```python
import matplotlib.pyplot as plt

def plot_history(history):
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  accuracy = history.history['accuracy']
  val_accuracy = history.history['val_accuracy']

  epochs = range(len(loss))

  fig, axs = plt.subplots(1, 2, sharex=True, squeeze=True, figsize=(12, 6))
  axs[0].plot(epochs, loss, label='loss')
  axs[0].plot(epochs, val_loss, label='val_loss')
  axs[0].set_title('loss')
  axs[0].set_xlabel('epochs')
  axs[0].legend()

  axs[1].plot(epochs, accuracy, label='accuracy')
  axs[1].plot(epochs, val_accuracy, label='val_accuracy')
  axs[1].set_title('accuracy')
  axs[1].set_xlabel('epochs')
  axs[1].legend()
```

```python
# Plot the loss and accuracy
plot_history(history=history_feature_extract)
```

Since we have a huge number of training datasets we can fine-tune by unfreezing **all** the layers of the base model. The training time should be very long. It could range from hours to 2-3 days as seen in DeepFood paper. 

ðŸ›  Due to the inherent nature of deep learning we don't know when the training will be finished so we have to setup a couple of more callbacks to monitor the training. 

```python
# Inspect the layers in saved gs model
for layer in load_gs_model.layers:
  print(layer.name, layer.trainable, layer.dtype, layer.dtype_policy)
```

```python
# Let's unfreeze all the base model's layers
for layer in load_gs_model.layers:
  layer.trainable = True
  print(layer.name, layer.trainable, layer.dtype, layer.dtype_policy)
```

```python
# Closely inspect the base model's layer
for layer in load_gs_model.layers[1].layers:
  if layer.trainable == True:
    continue
  print(layer.name, layer.trainable, layer.dtype, layer.dtype_policy)
```

OK so all the layers including the base model are now unfrozen i.e. trainable. 

ðŸ›  Let's set up callbacks now.

```python
# EarlyStopping callback
early_stopping_callback = tf.keras.callbacks.EarlyStopping(monitor="val_loss",
                                                           patience=3)

# ModelCheckpoint callback
checkpoint_path = "fine_tune_checkpoints/"
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path,
                                                               save_best_only=True,
                                                               monitor="val_loss")

# ReduceLROnPlateau callback
reduce_lr_callback = tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss",
                                                         factor=0.2,
                                                         patience=2,
                                                         verbose=1,
                                                         min_lr=1e-7)
```

```python
load_gs_model.compile(loss=tf.keras.losses.sparse_categorical_crossentropy,
                      optimizer=tf.keras.optimizers.Adam(0.0001),
                      metrics=['accuracy'])
```

```python
history_fine_tune = load_gs_model.fit(train_data,
                                      epochs=100,
                                      steps_per_epoch=len(train_data),
                                      validation_data=test_data,
                                      validation_steps=int(0.15 * len(test_data)),
                                      callbacks=[create_tensorboard_callback("training_logs", 
                                                                             "efficientnetb1_fine_tune"),
                                                 model_checkpoint_callback,
                                                 early_stopping_callback,
                                                 reduce_lr_callback])
```

```python
# Get the classification report and create DataFrame for class names against the f1 score 
from sklearn.metrics import classification_report
import pandas as pd

classification_report_dict = classification_report(actual_class, pred_class, output_dict=True)
class_f1_score = {}

for k, v in classification_report_dict.items():
  if k == 'accuracy':
    break
  class_f1_score[class_names[int(k)]] = v["f1-score"]

df_f1_score = pd.DataFrame({"class_name": list(class_f1_score.keys()),
                            "f1_score": list(class_f1_score.values())}).sort_values("f1_score", ascending=False)
df_f1_score
```

```python
# Plot the bar chart with f1 scores of different classes
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(15, 25))
f1_scores = axs.barh(range(len(df_f1_score)), df_f1_score['f1_score'].values)
axs.set_yticks(range(len(df_f1_score)))
axs.set_yticklabels(list(df_f1_score["class_name"]))
axs.set_xlabel("f1-score")
axs.set_title("F1-Scores")
axs.invert_yaxis(); # reverse the order

def autolabel(rects): # From: https://matplotlib.org/examples/api/barchart_demo.html
  for rect in rects:
    width = rect.get_width()
    axs.text(1.03*width, rect.get_y() + rect.get_height()/1.5,
            f"{width:.2f}",
            ha='center', va='bottom')

autolabel(f1_scores)
```

# Get the model's predictions on random test 

The goal is to perform following:
1. select 3 random images from the test data (images) from different classes
2. preprocess the images and feed into model to predict
3. get the predicted class, true class, prediction probability and top 3 predictions.

```python
# Create a definition to preprocess the images

def load_prep(filename, 
              img_shape=224,
              scale=True):
  img = tf.io.read_file(filename)
  img = tf.io.decode_image(img)
  img = tf.image.resize(img, [img_shape, img_shape])
  if scale:
    return img/255.
  return img
```

```python
# Make prediction on 3 random images
import os
import random


custom_images_path = "/content/drive/MyDrive/Food101_TransferLearning/Extras/Custom_images"

num_plots = 2
fig, axs = plt.subplots(num_plots, 2, figsize=(20, 15))
for i in range(num_plots):
    
    # Get the random image file path

    random_file = random.choice(os.listdir(custom_images_path))
    filepath = custom_images_path + "/" + random_file
    random_class_name = random_file[:-4]
    
    # Image preprocessing before it is fed inside model for prediction
    # Remember the model already has image rescaling layer built in so we don't need to rescale it
    img = load_prep(filepath, scale=False)
    img = tf.expand_dims(img, axis=0)
    
    # Predict the image using the model we built
    pred_prob_image = load_saved_fine_tune_model.predict(img)
    pred_class_image = class_names[pred_prob_image.argmax()]
    
    top_3_pred_prob_idx = (pred_prob_image.argsort())[0][-3:][::-1]
    top_3_pred_prob = [pred_prob_image[0][idx1] for idx1 in top_3_pred_prob_idx]
    top_3_pred_class = [class_names[idx2] for idx2 in top_3_pred_prob_idx]
    
    # Plot the figure
    # We need to remove that extra dimension related to the batch and divide the pixels by 255. for imshow() function
    axs[i, 0].imshow(tf.squeeze(img/255.))
    if random_class_name == pred_class_image:
        title_color = 'g'
    else:
        title_color = 'r'
    axs[i, 0].set_title(f"true label: {random_class_name} \n predicted label: {pred_class_image} \n probability: {pred_prob_image.max():.2f}", 
                     c=title_color)
    axs[i, 0].axis(False)
    
    axs[0, 1].set_title("Top 3 predictions")
    axs[i, 1].bar(top_3_pred_class, top_3_pred_prob);
    
```
